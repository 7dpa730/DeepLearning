```python
train_x=np.load("../input/kernel6d402bcb5a/onelakh_train_x.npz")
train_x=train_x['arr_0']
train_y=np.load("../input/kernel6d402bcb5a/onelakh_train_y.npz")
train_y=train_y['arr_0']
val_x=np.load("../input/kernel6d402bcb5a/fiftythousand_val_x.npz")
val_x=val_x['arr_0']
val_y=np.load("../input/kernel6d402bcb5a/fiftythousand_val_y.npz")
val_y=val_y['arr_0']
```

```python
train_y.shape
```

```python
def baseline_model():
# create model
    model = Sequential()
    model.add(Conv2D(32, (3, 3), input_shape=(1, 32, 32), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(32, (3, 3) ,activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Flatten())
    model.add(Dense(784, activation='sigmoid'))
    model.add(Dense(784, activation='sigmoid'))
    model.add(Dense(39, activation='softmax'))# Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python

# build the model
model = baseline_model()
# Fit the model
model.fit(train_x,train_y, validation_data=(val_x,val_y), epochs=30, batch_size=200, verbose=1)
# Final evaluation of the model
scores = model.evaluate(val_x,val_y, verbose=0)
print("CNN Error: %.2f%%" % (100-scores[1]*100))
```

```python
def le_net():
    lnm=Sequential()
    lnm.add(Conv2D(64, (3, 3), input_shape=(1, 32, 32), activation='relu'))
    lnm.add(MaxPooling2D(pool_size=(2, 2)))
    lnm.add(Conv2D(128,(3,3),activation='relu',padding='same'))
    lnm.add(Conv2D(256,(3,3),activation='relu'))
    lnm.add(AveragePooling2D(pool_size=(2, 2)))
    lnm.add(Dropout(0.2))
    lnm.add(Conv2D(512,(3,3),activation='relu',padding='same'))
    lnm.add(Conv2D(1024,(3,3),activation='relu'))
    lnm.add(AveragePooling2D(pool_size=(2, 2)))
    lnm.add(Dropout(0.4))
    lnm.add(Flatten())
    lnm.add(Dense(784, activation='relu'))
    lnm.add(Dense(400, activation='relu'))
    lnm.add(Dense(120, activation='relu'))
    lnm.add(Dense(39, activation='softmax'))
    lnm.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return lnm

    
```

```python
# build the model
lnm = le_net()
# Fit the model
lnm.fit(train_x,train_y, validation_data=(val_x,val_y), epochs=30, batch_size=100, verbose=1)
# Final evaluation of the model
scores = lnm.evaluate(val_x,val_y, verbose=0)
print("CNN Error: %.2f%%" % (100-scores[1]*100))
```
